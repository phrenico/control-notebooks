```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def hill_repression(x, n, k, alpha):
  return alpha / (1 + k * x**n)
```

```python
x = np.arange(0, 10, 0.1)
ns = np.arange(1, 5)
k = 1
alpha= 1

fs = [hill_repression(x, n, k, alpha) for n in ns]

fig = plt.figure()
_ = [plt.plot(x, f, label=str(ns[i])) for i, f in enumerate(fs)]
_ = plt.legend(title='Hill coef')
```

```python
def hill_activation(x, n, k, alpha):
  return (alpha * k * x**n) / (1 + k * x**n)

x = np.arange(0, 10, 0.1)
ns = np.arange(1, 5)
k = 1
alpha= 1

fs = [hill_activation(x, n, k, alpha) for n in ns]

fig = plt.figure()
_ = [plt.plot(x, f, label=str(ns[i])) for i, f in enumerate(fs)]
_ = plt.legend(title='Hill coef')
```
